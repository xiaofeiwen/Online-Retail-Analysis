```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
color = sns.color_palette()

%matplotlib inline
```

```python
# read data
all_data = pd.read_hdf('../data/online_retail.h5','clean')
orders = pd.read_hdf('../data/online_retail.h5','orders')
products = pd.read_hdf('../data/online_retail.h5','products')
customers = pd.read_hdf('../data/online_retail.h5','customers')
order_products = pd.read_hdf('../data/online_retail.h5','order_products')
product_features = pd.read_pickle('../data/product_features.p')
user_features = pd.read_pickle('../data/user_features.p')
```

# What are the number of orders per user?

```python
orders_by_users = orders.groupby("user_id")["order_number"].aggregate(np.max).reset_index()
max_orders = orders_by_users.order_number.max()
orders_by_users.loc[orders_by_users.order_number > 30,'order_number'] = 30
cnt_orders= orders_by_users.order_number.value_counts()

plt.figure(figsize=(12,8))
sns.barplot(cnt_orders.index, cnt_orders.values, alpha=0.8, color=color[0])
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('Number of Orders (Capped at 30)', fontsize=12)
plt.xticks(rotation='vertical')
plt.show()
print ('The maximum number of orders per user is {}.'.format(max_orders))
```

Number of invoices ranges from 4 to 202. Most of the users have less than 20 invoices.

# What are the number distinct products per orders?

```python
products_by_orders = order_products.groupby("order_id")["add_to_cart_order"].aggregate("max").reset_index()
max_products = products_by_orders.add_to_cart_order.max()
products_by_orders.loc[products_by_orders.add_to_cart_order > 60,'add_to_cart_order'] = 60
cnt_products = products_by_orders.add_to_cart_order.value_counts()

plt.figure(figsize=(16,8))
sns.barplot(cnt_products.index, cnt_products.values, alpha=0.8,color=color[1])
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('Number of distinct products per order (capped at 60)', fontsize=12)
plt.xticks(rotation='vertical')
plt.show()
print ('The maximum number of distinct products per order is {}.'.format(max_products))
```

```python
products_by_orders = order_products.groupby("order_id")["add_to_cart_order"].aggregate("max").reset_index()
print ("The average number of distinct products per transaction was {:.0f};\n"
        "The average quantity of products per transaction was {:.0f}."
      .format(products_by_orders.add_to_cart_order.mean(),all_data.groupby('order_id')['quantity'].sum().mean()))
```

This seems to suggest that many of the consumers of the business were organizational customers rather than individual customers.

# Are there outlier users in terms of how much they bought and average price they paid?

```python
plt.figure(figsize=(16,8))
plt.scatter(x='avg_money_spent_per_item',y='total_item_quantity',data=user_features)
plt.xlabel('average money spent per item')
plt.ylabel('total item quantity')
plt.show()
```

There is one users buying large quantity of cheap products. That users might be a reseller of cheap gifts.

# When do users place their orders?

```python
fig, ax = plt.subplots(figsize=(12,8))
sns.countplot(x="order_dow", data=orders, color=color[0])
plt.ylabel('Count', fontsize=12)
plt.xlabel('', fontsize=12)
locs, labels = plt.xticks()
plt.title("Frequency of order by day of week", fontsize=15)

labels = [item.get_text() for item in ax.get_xticklabels()]
labels = ['Monday','Tuesday','Wednesday','Thursday','Friday','Sunday']

ax.set_xticklabels(labels)

plt.show()
```

The amount of orders placed increases from Monday to Thursday and decreases from Thursday to Sunday. Strangely, no orders were placed on Saturday.

```python
plt.figure(figsize=(12,8))
sns.countplot(x="order_hour_of_day", data=orders, color=color[1])
plt.ylabel('Count', fontsize=12)
plt.xlabel('Hour of day', fontsize=12)
plt.xticks(rotation='vertical')
plt.title("Frequency of order by hour of day", fontsize=15)
plt.show()
```

The majority of orders are placed during business hours. The amount of orders placed increases from 7am to 12pm and decreases from 12pm to 20pm. No orders were placed between 20pm and 7am.

```python
grouped_df = orders.groupby(["order_dow", "order_hour_of_day"])["order_id"].agg("count").reset_index()
grouped_df = grouped_df.pivot('order_dow', 'order_hour_of_day', 'order_id')

fig, ax = plt.subplots(figsize=(12,8))
sns.heatmap(grouped_df)
plt.title("Frequency of Day of week Vs Hour of day")
plt.xlabel('order hour of day')
plt.ylabel('')
labels = [item.get_text() for item in ax.get_yticklabels()]
labels = reversed(['Monday','Tuesday','Wednesday','Thursday','Friday','Sunday'])
ax.set_yticklabels(labels)

plt.show()
```

Wednesdays and Thursdays at noon has the largest amount of order flow.

# Most ordered products

```python
def top_10_product_features(column):
    df = product_features.sort_values(column,ascending=False).iloc[:10]
    plt.figure(figsize=(16,8))
    sns.barplot(df.description.str.lower(), df[column], alpha=0.8,color=color[1])
    plt.ylabel(column, fontsize=16)
    plt.xlabel('')
    plt.xticks(rotation=60)
    plt.show()
top_10_product_features('orders')
```

```python
def showImagesHorizontally(folderPath):
    filenames = os.listdir(folderPath)
    list_of_files = [os.path.join(folderPath,filename) for filename in filenames]
    list_of_files.sort()
    fig = plt.figure(figsize=(16,8))
    number_of_files = len(list_of_files)
    for i in range(number_of_files):
        a=fig.add_subplot(1,number_of_files,i+1)
        image = plt.imread(list_of_files[i])
        plt.imshow(image,cmap='Greys_r')
        plt.axis('off')
showImagesHorizontally('../image/most ordered')
plt.text(-6000, -100, 'Most ordered products',size=15);
```

# What are the reorder behaviors for users from different countries?

```python
df = customers.set_index('user_id').join(user_features)
df = df[df.country.isin(['United Kingdom','Germany','France'])]
plt.figure(figsize=(12,8))
sns.violinplot(x='country',y='user_reorder_ratio',data=df)
plt.xlabel('')
plt.ylabel('user reorder ratio')
plt.show()
```

Users from France are slightly more likely to reorder than users from Germany and United Kindom.
